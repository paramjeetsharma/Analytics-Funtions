```python
import pandas as pd #Python data analysis library
import numpy as np #Python scientific computing
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
from google.colab import drive
drive.mount('/gdrive')
```

```python

df = pd.read_csv('/gdrive/My Drive/Online.csv')
```

```python
df.head()
```

```python
productCountDf = df.groupby("Product",as_index = False)["Transaction ID"].count()
```

```python
productCountDf
```

```python
#Top 100 most frequently purchased products
topLev = 50

#Here order_id is the count so we need to sort the data frame w.r.t order_id
productCountDf = productCountDf.sort_values("Transaction ID",ascending = False)

topProdFrame = productCountDf.iloc[0:topLev,:]
```

```python
topProdFrame
```

```python
df1 = df[0:0]
for i in range(0,49):
    pId = topProdFrame.iloc[i]['Product'] 
    stDf = df[df.Product == pId ]
    df1 = df1.append(stDf,ignore_index = False)
```

```python
df1.head()
```

```python
df2=df1[['Product','Transaction ID','Quantity']]
```

```python

basket = pd.pivot_table(data=df2,index='Transaction ID',columns='Product',values='Quantity', \
                        aggfunc='sum',fill_value=0)
```

```python
# Convert the units to 1 hot encoded values
def encode_units(x):
    if x == 0:
        return 0
    else:
        return 1
```

```python

basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python

basket_sets.size
```

```python
# Build up the frequent items

frequent_itemsets = apriori(basket_sets,  min_support=0.001 ,   use_colnames=True)
```

```python
#frequent_itemsets=fpgrowth(basket_sets, min_support=0.6, use_colnames=True)
```

```python
frequent_itemsets
```

```python

rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_mlxtend
```
